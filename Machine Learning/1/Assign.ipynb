```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
data = pd.read_csv("/content/dataset.csv")
data.head(5)
```

```python
x = data.iloc[:,0:6]
y= data.iloc[:,6:7]
```

```python
#split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors =3)
model.fit(x_train,y_train)
```

```python
print(model.score(x_train,y_train)*100)
print(model.score(x_test,y_test)*100)
```

#KNN WEIGHTED

```python
model2 = KNeighborsClassifier(n_neighbors=5,metric='euclidean',weights='distance')
model2.fit(x_train,y_train)
```

```python
print(model2.score(x_train,y_train)*100)
print(model2.score(x_test,y_test)*100)
```

##Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model3 = GaussianNB()
model3.fit(x_train,y_train)
```

```python
print(model3.score(x_train,y_train)*100)
print(model3.score(x_test,y_test)*100)
```

```python
from sklearn.naive_bayes import BernoulliNB
model4 = BernoulliNB()
model4.fit(x_train,y_train)
```

```python
print(model4.score(x_train,y_train)*100)
print(model4.score(x_test,y_test)*100)
```
